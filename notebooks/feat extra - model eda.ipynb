# LightGBM Exploration

## 1. Library Imports

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import optuna
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import balanced_accuracy_score, log_loss, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore', category=UserWarning)
```

## 2. Configuration

```python
DATA_PATH = "../data/processed/train_engineered_features.csv"

TARGET_COL = 'target'
ID_COL = 'ID'

SEED = 42
N_FOLDS = 5
N_OPTUNA_TRIALS = 50
OPTUNA_TIMEOUT = 600
```

## 3. Data Loading

```python
try:
    df = pd.read_csv(DATA_PATH)
    print(f"Data loaded from: {DATA_PATH}")
    print(f"Data shape: {df.shape}")
except FileNotFoundError:
    print(f"ERROR: File {DATA_PATH} not found.")
    print("Check the path or run the feature_pipeline.py pipeline first.")
    df = None
```

## 4. Quick Exploration and Preparation

```python
if df is not None:
    print("\n--- Data Information ---")
    df.info()

    print("\n--- First rows ---")
    print(df.head())

    print("\n--- Missing values per column ---")
    print(df.isnull().sum().sort_values(ascending=False))

    if TARGET_COL in df.columns:
        y = df[TARGET_COL]

        if not pd.api.types.is_numeric_dtype(y):
            print(f"\nEncoding target column '{TARGET_COL}'...")
            le = LabelEncoder()
            y = le.fit_transform(y)
            print("Encoded classes:", le.classes_)

        feature_columns = df.columns.drop([ID_COL, TARGET_COL])
        X = df[feature_columns]
        print(f"\nShape of X (features): {X.shape}")
        print(f"Shape of y (target): {y.shape}")

        print("\n--- Target Distribution ---")
        target_dist = pd.Series(y).value_counts(normalize=True).sort_index()
        print(target_dist)

        plt.figure(figsize=(6, 4))
        sns.countplot(x=y)
        plt.title('Target Variable Distribution')
        plt.xlabel('Target Class')
        plt.ylabel('Number of samples')

        if 'le' in locals():
             plt.xticks(ticks=range(len(le.classes_)), labels=le.classes_)
        plt.show()

    else:
        print(f"ERROR: Target column '{TARGET_COL}' not found in DataFrame.")
        X, y = None, None
```

## 5. Hyperparameter Optimization with Optuna and LightGBM

```python
def objective(trial, X, y, n_folds=N_FOLDS, random_state=SEED):
    """Objective function for Optuna optimization with LGBMClassifier."""

    params = {
        'objective': 'multiclass',
        'metric': 'multi_logloss',
        'num_class': len(np.unique(y)),
        'boosting_type': 'gbdt',
        'n_estimators': trial.suggest_int('n_estimators', 100, 2000, step=100),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 20, 300),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.4, 1.0),
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.4, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
        'lambda_l1': trial.suggest_float('lambda_l1', 1e-8, 10.0, log=True),
        'lambda_l2': trial.suggest_float('lambda_l2', 1e-8, 10.0, log=True),
        'class_weight': trial.suggest_categorical('class_weight', [None, 'balanced']),
        'verbose': -1,
        'n_jobs': -1,
        'seed': random_state
    }

    cv_scores = []
    cv = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=random_state)

    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]

        model = lgb.LGBMClassifier(**params)
        model.fit(X_train, y_train,
                  eval_set=[(X_val, y_val)],
                  eval_metric='multi_logloss',
                  callbacks=[lgb.early_stopping(100, verbose=False)])

        preds = model.predict(X_val)
        score = balanced_accuracy_score(y_val, preds)
        cv_scores.append(score)

    return np.mean(cv_scores)


if X is not None and y is not None:
    print(f"\n--- Starting Optuna optimization ({N_OPTUNA_TRIALS} trials) ---")
    study = optuna.create_study(direction='maximize', study_name='LGBM Optimization')

    func = lambda trial: objective(trial, X, y)
    study.optimize(func, n_trials=N_OPTUNA_TRIALS, timeout=OPTUNA_TIMEOUT, show_progress_bar=True)

    print("\n--- Best Optimization Results ---")
    print(f"Number of completed trials: {len(study.trials)}")
    print(f"Best value (Mean CV Balanced Accuracy): {study.best_value:.4f}")
    print("Best hyperparameters found:")
    best_params = study.best_params
    print(best_params)
else:
    print("\nOptuna optimization canceled because X or y data could not be prepared.")
    best_params = {}
```

## 6. Final Model Training and Evaluation

```python
if best_params and X is not None and y is not None:
    print("\n--- Training and Evaluation with Best Hyperparameters (Cross-Validation) ---")

    final_params = best_params.copy()
    final_params.update({
        'objective': 'multiclass',
        'metric': 'multi_logloss',
        'num_class': len(np.unique(y)),
        'boosting_type': 'gbdt',
        'verbose': -1,
        'n_jobs': -1,
        'seed': SEED
    })

    if 'n_estimators' in final_params:
        print(f"Note: 'n_estimators' suggested by Optuna ({final_params['n_estimators']}) will potentially be adjusted by early stopping.")

        final_params['n_estimators'] = final_params.get('n_estimators', 1000)


    cv = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=SEED)
    oof_preds = np.zeros(len(X))
    oof_proba = np.zeros((len(X), len(np.unique(y))))
    models = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y)):
        print(f"\n--- Fold {fold+1}/{N_FOLDS} ---")
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]

        model = lgb.LGBMClassifier(**final_params)
        model.fit(X_train, y_train,
                  eval_set=[(X_val, y_val)],
                  eval_metric='multi_logloss',
                  callbacks=[lgb.early_stopping(100, verbose=True)])

        val_preds = model.predict(X_val)
        val_proba = model.predict_proba(X_val)
        oof_preds[val_idx] = val_preds
        oof_proba[val_idx, :] = val_proba
        models.append(model)

        fold_score = balanced_accuracy_score(y_val, val_preds)
        print(f"Fold {fold+1} Balanced Accuracy: {fold_score:.4f}")

    print("\n--- Overall Evaluation (Out-of-Fold) ---")
    oof_balanced_accuracy = balanced_accuracy_score(y, oof_preds)

    try:
        oof_logloss = log_loss(y, oof_proba)
        print(f"OOF Balanced Accuracy : {oof_balanced_accuracy:.4f}")
        print(f"OOF LogLoss           : {oof_logloss:.4f}")
    except ValueError:
        print(f"OOF Balanced Accuracy : {oof_balanced_accuracy:.4f}")
        print("OOF LogLoss           : N/A (Calculation Error)")


    # Classification Report
    print("\n--- Classification Report (OOF) ---")
    target_names = [f'Class {i}' for i in np.unique(y)]
    if 'le' in locals():
        target_names = [f'Class {c}' for c in le.classes_]
    print(classification_report(y, oof_preds, target_names=target_names))

    # Confusion Matrix
    print("\n--- Confusion Matrix (OOF) ---")
    cm = confusion_matrix(y, oof_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
    fig, ax = plt.subplots(figsize=(8, 6))
    disp.plot(ax=ax, cmap=plt.cm.Blues)
    plt.title('Confusion Matrix (Out-of-Fold Predictions)')
    plt.show()
```

## 7. Feature Importance

```python
if best_params and X is not None and y is not None:
    print("\n--- Feature Importance (Model trained on all X) ---")
    final_model_for_importance = lgb.LGBMClassifier(**final_params)
    final_model_for_importance.fit(X, y)

    lgb.plot_importance(final_model_for_importance, max_num_features=20, figsize=(10, 8))
    plt.title('Top 20 Feature Importances (LGBM)')
    plt.tight_layout()
    plt.show()

    import joblib
    model_filename = "../models_trained/classic/lgbm_final_model.pkl"
    joblib.dump(final_model_for_importance, model_filename)
    print(f"\nFinal model saved to {model_filename}")

elif df is not None:
    print("\nFinal model training canceled because optimization did not provide parameters.")
else:
    print("\nAnalysis canceled because data could not be loaded.")
```

## 8. Generate Predictions on Test Data

```python
import joblib
from pathlib import Path

print("\n--- Generating Predictions on Test Set ---")

MODEL_PATH = Path("../models_trained/classic/lgbm_final_model.pkl")
TEST_DATA_PATH = Path("../data/processed/test_engineered_features.csv") # Use the filtered test data
SUBMISSION_PATH = Path("../submissions/lgbm_submission.csv") # Optional: Define submission file path

SUBMISSION_PATH.parent.mkdir(parents=True, exist_ok=True)

try:
    loaded_model = joblib.load(MODEL_PATH)
    print(f"Model loaded successfully from: {MODEL_PATH}")
except FileNotFoundError:
    print(f"ERROR: Model file not found at {MODEL_PATH}")
    print("Please ensure the model was trained and saved correctly in the specified path.")
    loaded_model = None
except Exception as e:
    print(f"Error loading model: {e}")
    loaded_model = None

if loaded_model:
    try:
        df_test = pd.read_csv(TEST_DATA_PATH)
        print(f"\nTest data loaded successfully from: {TEST_DATA_PATH}")
        print(f"Test data shape: {df_test.shape}")
    except FileNotFoundError:
        print(f"ERROR: Test data file not found at {TEST_DATA_PATH}")
        print("Please ensure the test data processing pipeline was run.")
        df_test = None
    except Exception as e:
        print(f"Error loading test data: {e}")
        df_test = None

if loaded_model and df_test is not None:
    if ID_COL not in df_test.columns:
        print(f"ERROR: ID column '{ID_COL}' not found in test data.")
    else:
        test_ids = df_test[ID_COL]

        if 'X' in locals() and isinstance(X, pd.DataFrame):
            feature_columns_for_test = X.columns
            print(f"\nUsing {len(feature_columns_for_test)} feature columns identified during training.")

            missing_cols = [col for col in feature_columns_for_test if col not in df_test.columns]
            if missing_cols:
                print(f"ERROR: The following feature columns used for training are missing in the test data:")
                print(missing_cols)
                X_test = None
            else:
                 X_test = df_test[feature_columns_for_test]

                 if X_test.isnull().sum().sum() > 0:
                     print(f"\nWarning: Found {X_test.isnull().sum().sum()} NaN values in test features. Imputing with 0.")
                     X_test = X_test.fillna(0.0)

                 X_test.replace([np.inf, -np.inf], 0.0, inplace=True)

                 print("\nMaking predictions on the test set...")
                 try:
                     predictions = loaded_model.predict(X_test)
                     print("Predictions generated.")

                     print("\nCreating submission DataFrame...")
                     submission_df = pd.DataFrame({ID_COL: test_ids, TARGET_COL: predictions})

                     print("Sorting submission DataFrame by ID...")
                     submission_df = submission_df.sort_values(by=ID_COL, ascending=True)

                     print("\n--- Submission DataFrame (Head) ---")
                     print(submission_df.head())

                     try:
                         submission_df.to_csv(SUBMISSION_PATH, index=False)
                         print(f"\nSubmission file saved successfully to: {SUBMISSION_PATH}")
                     except Exception as e:
                         print(f"\nError saving submission file: {e}")

                 except Exception as e:
                     print(f"Error during prediction or submission file creation: {e}")

        else:
             print("ERROR: Training features DataFrame 'X' not found in memory. Cannot determine test features.")

print("\nPrediction cell finished.")
```
